
# 일차 목표 

일단 각 대분류당 이미지를 제대로 검출하는지 확인한다. 



```python
def func_dir(home_path, adr):
    """
    home_path(원하는 경로) = 'C:\\split_data' 
    adr = 파일 이름
    디렉토리 만들기 -> adr에 파일 이름만 추가 하면 됨 
    """
    base = home_path # 파일 경로 
    dir_path = os.path.abspath(home_path) # 경로 
#     base = os.getcwd()
#     full = base + adr
    tmp = adr.split('\\')

    for idx in range(len(tmp)):
        ck = base +'\\'+'\\'.join(tmp[:idx+1])
        if os.path.exists(ck) is False:
            os.mkdir(ck)
            print('데이터 확인 :', ck , os.path.isdir(ck))
```

```python
home_path = 'C:\\split_data' # 파일 경로 
dir_path = os.path.abspath(home_path) # 경로 

class_num = list(range(0,150))
set_type = ['train', 'test','valid']

print(dir_path)
```
